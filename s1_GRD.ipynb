# Exploring S1 GRD image with `xarray sentinel`

Example notebook is [here](https://github.com/bopen/xarray-sentinel/blob/main/notebooks/Sentinel-1_GRD_IW.ipynb)

```python
import xarray as xr
import xarray_sentinel
import matplotlib.pyplot as plt
```

```python
product_path = ('S1A_IW_GRDH_1SDV_20220430T000314_20220430T000339_042995_05221E_C8B3.SAFE')

swath_group = 'IW'
swath_polarisation_group = 'IW/VV'
measurement_group = 'IW/VV'
measurement_block_slices = (slice(1000, 9000), slice(20000, 30000))
digital_number_max = 600
```

```python
ds = xr.open_dataset(product_path, engine='sentinel-1')
```

```python
ds
```

```python
ds1 = xr.open_dataset(product_path, engine='sentinel-1', group=swath_polarisation_group)
ds1
```

```python
measurement = xr.open_dataset(
    product_path, engine="sentinel-1", group=measurement_group, chunks=2048
)
measurement
```

```python
measurement_block = measurement.measurement[measurement_block_slices]
measurement_block
```

```python
_ = abs(measurement.measurement).plot(y="azimuth_time", vmax=digital_number_max)
```
